```python
with open("../raw_results/omcs/filtered/locations_filtered.txt", 'r') as file3:
    locs = [line.split(':')[0] for line in file3.readlines()]


for loc in locs[:10]:
    print(loc)
```

### Warning: The code below will take up significant computing resources and generate a large file (~4GB)

```python
import json

with open("tmp_compositional_sents.json", 'w') as out:
    sent_list = []
    i = 0
    for sub in subjs:
        for loc in locs:
            for vp in vps:
                local = {}
                # who + doing what + at where
                local["id"] = i
                local["subj"] = sub
                local["vp"] = vp
                local["loc"] = loc
                local["text"] = sub + ' ' + vp + ' ' + loc
                # print(local["text"])
                sent_list.append(local)
                i += 1
                if (i % 100 == 0):
                    print(f"Processed {i} sentences")
    print(f"Processed {i} sentences")
    j_f = {'data': sent_list}
    json.dump(j_f, out)
```
